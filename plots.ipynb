```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
%matplotlib inline

```

```python
text = """
Eval perplexity:  11468.40690239118
lr: 0.000017, step: 1000, loss: 2.633574
Eval perplexity:  20893.139980762608
lr: 0.000015, step: 2000, loss: 2.167759
Eval perplexity:  110130.31652505911
lr: 0.000012, step: 3000, loss: 1.787407
Eval perplexity:  300898.27128823835
lr: 0.000009, step: 4000, loss: 1.491481
Eval perplexity:  1149326.933106448
lr: 0.000006, step: 5000, loss: 1.261684
Eval perplexity:  1602050.5018750026
lr: 0.000003, step: 6000, loss: 1.072305
Eval perplexity:  1944245.1525715312
lr: 0.000000, step: 7000, loss: 0.972159
"""
```

```python
ppls = []
losses = []

for line in text.splitlines():
    if "Eval perplexity" in line:
        ppls.append(float(line.split(":")[-1]))
    if "loss" in line:
        losses.append(float(line.split(":")[-1]))
```

```python
# plot perplexity and loss on the same plot, with different y-axis
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(ppls, color="red")
ax1.set_xlabel("step")
ax1.set_ylabel("perplexity", color="red")
ax1.tick_params(axis="y", labelcolor="red")

ax2 = ax1.twinx()
ax2.plot(losses, color="blue")
ax2.set_ylabel("loss", color="blue")
ax2.tick_params(axis="y", labelcolor="blue")
```
